```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pitch_df = pd.read_csv('../Data/Alap/AG_1a_Jaun/AG_1a_Jaun-pitch.csv')
```

```python
pitch_vals = pitch_df.loc[pitch_df['time'] < 12, 'pitch'].values
pitch_vals[pitch_vals == -3000] = np.nan
time_vals = pitch_df.loc[pitch_df['time'] < 12, 'time'].values
```

```python
plt.plot(time_vals, pitch_vals)
plt.title('Pitch Contour')
plt.xlabel('Time (s)')
plt.ylabel('Cents')
plt.xlim((-0.5, 12.5))
plt.savefig('/home/nithya/Desktop/pitch.png')
```
