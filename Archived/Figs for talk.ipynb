```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
plt.style.use('ggplot')
```

```python
file = '../Data/Alap/AG_4a_Nand/AG_4a_Nand-pitch.csv'
pitch = pd.read_csv(file)
pitch = pitch.loc[(pitch['time'] >= 3.5) & (pitch['time'] <= 10.5)]
```

```python
pitch.loc[pitch['pitch'] == -3000] = np.nan
```

```python
plt.plot(pitch['time'], pitch['pitch'])
plt.yticks([-300, -100, 0, 200, 400, 500, 700], ['D_', 'N_', 'S', 'R', 'G', 'M', 'P'])
plt.xticks([])
plt.xlabel('Time')
plt.ylabel('Notes')
plt.savefig('../Presentation/pitch.png')
```
