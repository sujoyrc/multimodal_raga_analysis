```python
import numpy as np
import pandas as pd
import os
import matplotlib
import matplotlib.pyplot as plt
```

```python
file_name_1 = 'Alap/AG_4a_Nand.csv'
file_name_2 = 'Alap/SCh_3c_MM.csv'
file_name_3 = 'Pakad/CC_P6_Kedar.csv'

file_to_plot = file_name_3 # pick file to plot
```

```python
orig_folder = '../Final Video Data/CSV_solo_unfiltered/'
replaced_folder = '../Final Video Data/Replaced_nan/'
filter_folder = '../Final Video Data/Filter_values/'
normalized_folder = '../Final Video Data/Normalized/'
```

```python
cmap = plt.get_cmap('tab20')
cs = [cmap(i) for i in np.linspace(0, 1, 13)]
```

### Plot original curves

```python
orig_df = pd.read_csv(os.path.join(orig_folder, file_to_plot), header=[0, 1])
```

```python
orig_df.head()
```

```python
# there are 5 nan values in RWrist x, y and c variables
orig_df[orig_df['RWrist', 'x'].isnull() == True]['RWrist', 'x']
```

```python
# plot x coords
plt.figure(figsize=(14, 10))
i=0
for cols in orig_df.columns:
    if cols[1]=='x':
        plt.plot(np.arange(orig_df.shape[0]), orig_df[cols], label=cols[0], color=cs[i])
        i += 1
plt.legend()
```

```python
# plot y coords
plt.figure(figsize=(14, 10))
i=0
for cols in orig_df.columns:
    if cols[1]=='y':
        plt.plot(np.arange(orig_df.shape[0]), orig_df[cols], label=cols[0], color=cs[i])
        i += 1
plt.legend()
```
